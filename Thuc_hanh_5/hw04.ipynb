```python
# W4A18 : Nhập vào số nguyên dương a và b, in toàn bộ ước chung của a và b
a,b=map(int,input().split())
for i in range(1,min(a,b) +1):
    if a%i==0 and b%i==0 :
        print(i,end=" ")
# Phép thử a,b=20,6
```

```python
# W4A20 Nhập vào một số nguyên dương n, kiểm tra xem n có phải là số dạng 2^k hay không 
n = int(input())
for i in range(1,n+1) :
    if n==2**i :
        print(n,"có dạng là 2**{0}".format(i))
        break
else :
    print(n,"không có dạng 2**k")
```

```python
# W4A21 Nhập vào số nguyên dương n, tính tổng các chữ số của n 
n = int(input())
tong = 0
while n>0 :
    s = n%10
    tong+=s
    n=n//10
print(tong)
# Phép thử n = 27102007
    
    
```

```python
# W4A22 Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số chẵn, bao nhiêu chữ số lẻ 
n = int(input())
x = n
so_chan = 0
so_le = 0
while n>0 :
    s = n%10
    if s%2==0 :
        so_chan+=1
    else :
        so_le+=1
    n = n//10
print(x,"có số chẵn là : ",so_chan)
print(x,"có số lẻ là : ",so_le)
# Phép thử n = 2468
```

```python
""" W4A23 : Nhập n 
Cho S(k) = 1 + 2 + 3 + … + k 
Tìm k sao cho S(k) lớn nhất nhưng nhỏ hơn n 
"""
n = int(input())
S = 0
k = 0
while (S + (k+1)) < n :
    k+=1
    S+=k
print("{0} thỏa mãn tổng {1} lớn nhất nhưng bé hơn {2} ".format(k,S,n))
  
```

```python
""" W4A24 : Nhập vào A, tìm n nhỏ nhất sao cho 
1 + 1/2 + 1/3 + 1/4 + ... + 1/n > A
"""
A = int(input())
s = 0 
n = 1
while A > 0 :
    s = s + 1/n
    n = n +1 
    if s>A :
        break
print(n)
        
# Phép thử A = 10
```

```python
""" W4A25 : Nhập vào một dãy số nguyên, ngưng nhập khi người dùng nhập -1. 
Sau khi nhập xong, in số lớn nhất, số nhỏ nhất trong những số vừa nhập 
"""
str1 = []
while True :
    n = int(input())
    if n == -1:
        break
    str1.append(n)
if len(str1) == 0 :
    print("Không có số nào được nhập ")
else :
    print("Số lớn nhất : {0} và số bé nhất là : {1}".format(max(str1),min(str1)))        
        
```

```python
""" W4A26 : Dãy số fibonacci là dãy số được định nghĩa như sau: 1, 1, 2, 3, 5, 8, 13,... với số kế tiếp sẽ 
bằng tổng hai số trước đó 
Nhập vào số nguyên dương A, hãy tìm số trong dãy số fibonacci lớn nhất nhưng không vượt 
quá A 
"""
A = int(input())
f1 = 1
f2 = 1
while f1 + f2 <=A :
    fn = f1 +f2
    f1 = f2
    f2 = fn
print(f2)
# Phép thử A = 10
```

```python
""" W4A27 : Nhập vào một chuỗi, hãy đếm xem trong chuỗi có bao nhiêu từ (quy định là chuỗi không có 
ký tự đặc biệt, không số, không có dấu câu, chỉ có ký tự chữ và khoảng trắng) 
Dùng cách: for trên index; KHÔNG dùng hàm split(), len() 
"""
n =input()
tong = 0
for  i in n :
    tong+=1 
print(tong)
    
```

```python
""" W4A28 : Nhập vào một chuỗi, hãy in từ đầu tiên trong chuỗi 
Dùng cách: for trên index; KHÔNG dùng hàm split() 
"""
s = input()
word = ''
for i in range(len(s)) :
    if s[i]=='' :
        break
    word += s[i]
print(word)
```

```python
""" W4A31 : Nhập vào một chuỗi, hãy đếm xem trong chuỗi có bao nhiêu ký tự in hoa, bao nhiêu ký tự in 
thường, bao nhiêu ký tự số"""
str1 = input()
in_hoa = 0
in_thuong = 0
ki_tu_so = 0
for c in str1 :
    if c.isupper() :
        in_hoa+=1
    elif c.islower() :
        in_thuong+=1
    elif c.isdigit() :
        ki_tu_so+=1
print("Ký tự in hoa là : ",in_hoa) 
print("Ký tự in thường là : ",in_thuong)
print("Ký tự số là : ",ki_tu_so)      
# Phép thử str1 = NguyenHaNinTon2007
```
