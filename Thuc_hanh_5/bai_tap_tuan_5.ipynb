```python
# W5A1 : Viết hàm truyền vào hai số nguyên a, b. Trả về số lớn hơn trong hai số. def Max_Of_Two( a, b) -> int:
def max_ab(a,b) :
    if a>b :
        return a 
    else :
        return b 
a,b = map(int,input().split())
print(max_ab(a,b))
```

```python
# W5A2 : Viết hàm hoán đổi hai số nguyên. def swap(a, b) -> Tuple[int, int]:  
def swap(a,b) :
    temp = a
    a = b
    b = temp
    return a,b
a,b = map(int,input().split())
print(swap(a,b))
```

```python
# W5A3 : Viết hàm kiểm tra một số có phải là số nguyên tố hay không?
def so_nguyen_to(n) :
    for  i in range(2,n) :
        if n%i==0 :
            return False 
    else :
        return True
n = int(input())
print(so_nguyen_to(n))
```

```python
# W5A13 : Viết hàm max_of_two(a, b) trả về số lớn hơn trong hai số. 
def max_of_two(a,b) :
    if a>b :
        return a
    elif b>a :
        return b 
    else :
        return a == b 
a,b = map(int,input().split())
print(max_of_two(a,b))
```

```python
# W5A14 : Viết hàm compute(a, b, op) thực hiện phép tính a (+|-|*|/) b tùy theo ký tự op. 
def compute(a,b,op) :
    if op == '/' :
        return a/b
    elif op == '*' :
        return a*b
    elif op == '-' :
        return a-b 
    elif op == '+' :
        return a+b 
op = input()
a,b = map(int,input().split())
print(compute(a,b,op))
```

```python
# W5A15 : Viết hàm is_right_triangle(a, b, c) trả về True nếu ba số có thể là ba cạnh của tam giác vuông. 
def is_right_triangle(a,b,c) :
    if a**2==b**2 + c**2 or b**2==a**2 + c**2 or c**2 == b**2 + a**2 :
        return True
    else :
        return False
a,b,c = map(int,input().split())
print(is_right_triangle(a,b,c))
```
