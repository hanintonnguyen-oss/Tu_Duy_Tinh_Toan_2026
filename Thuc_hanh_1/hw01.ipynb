```python
# Bài 1: Nhập vào số nguyên n, in ra màn hình số gấp đôi của n 
n = int(input("Nhập vào số nguyên n : "))
print("Số gấp đôi của", n, "là", 2 * n)
```

```python
""" Bài 2: Một khu đất hình chữ nhật có chiều rộng a, chiều dài b. Người ra xây một khu 
vui chơi hình tròn tại vị trí trung tâm khu đất như hình vẽ. Nhập vào a, b. Hãy tính phần 
diện tích còn để trồng cây xung quanh là bao nhiêu. Biết pi = 3.14, lấy kết quả đến 2 
chữ số phần thập phân.
"""
a,b = map(float,input("Nhập vào chiều dài a và chiều rộng b của khu đất").split())
pi = 3.14
r = min(a,b)/2
s1 = a*b
s2 = pi*r**2
s = s1 - s2
print("Phần diện tích trồng cây cung quanh khu vui chơi là : ",round(s,2))
```

```python
""" Bài 3: Viết chương trình nhập vào một “kí tự” c, nếu nó là chữ hoa thì in ra chữ thường 
tương ứng và ngược lại nếu nó là chữ thường thì in ra chữ hoa tương ứng của nó. (xem 
trước cấu trúc lựa chọn: if….else, ma) 
"""
c = input("Nhập vào một kí tự : ")
if c.isupper() :
    print("Chữ thường tươn ứng của ",c,"là",c.lower())
else:
    print("Chữ hoa tương ứng của ",c,"là",c.upper())
```

```python
""" Bài 4: Viết chương trình nhập vào một “kí tự”, Kiểm tra xem c có phải là kí tự chữ cái 
không (hoa hoặc thường). Nếu đúng thì in ra “{c} là kí tu alphabet”, nếu sai in ra “{c} 
không phải là kí tự alphabet. 
"""
c = input("Nhập vào một kí tự : ")
if c.isalpha():
    print(f"{c} là kí tự alphabet")
else:
    print(f"{c} không phải là kí tự alphabet")
```

```python
""" Bài 5: Nhập vào chữ cái hoa A, in ra chữ cái thường liền trước chữ cái a thường tương 
ứng của A. Xét TH khi c = ‘A’ là đặc biệt. 
"""
c = input("Nhập vào một chữ cái hoa : ")
thuong = c.lower()
if thuong == 'a' :
    print('z')
else :
    print(chr(ord(thuong)-1))
```

```python
""" Bài 6. Nhập vào 3 số a, b, c. Kiểm tra xem đây có phải là cạnh của 1 tam giác không, 
nếu đúng thì tính diện tích của tam giác đó, in ra với 1 chữ số phần thập phân, nếu nó 
không phải 3 cạnh của tam giác thì in ra “Khong phai 3 canh cua tam giac”. 
"""
a,b,c = map(float,input("Nhập vào 3 cạnh của tam giác : ").split())
if a+b>c and a+c>b and b+c>a :
    print(a,b,c,"Đây là ba cạnh của tam giác ")
    p = (a+b+c)/2
    s = (p*(p-a)*(p-b)*(p-c))**0.5
    print("Diện tích tam giác có độ dài cạnh", a, b, c, "là", round(s, 1))
else :
    print(a,b,c,"Không phải 3 cạnh của tam giác")
```

```python
""" Bài 7: Nhập vào một chuỗi gồm các kí tự thường độ dài >=20, in ra chữ cái thứ năm và 
thứ chín của chuỗi này. 
"""
c = input("Nhập vào một chuỗi gồm các kí tự thường độ dài >=20 : ")
if len(c)>=20 :
    print("Chữ cái thứ 5 của chuỗi là : ",c[4])
    print("Chữ cái thứ 9 của chuỗi là : ",c[8])
```

```python
""" Bài 8: Nhập vào họ tên chủ hộ và chỉ số điện kế tháng trước và chỉ số điện kế tháng này, 
tính tiền điện tháng này cho hộ biết:  - 
Từ 0 – 50kWh: 1.984 đồng/kWh  - - - - - 
Bậc 2: Từ 51 – 100 kWh: 2.050 đồng/kWh  
Bậc 3: Từ 101 – 200 kWh: 2.380 đồng/kWh  
Bậc 4: Từ 201 – 300 kWh: 2.998 đồng/kWh  
Bậc 5: Từ 301 – 400 kWh: 3.350 đồng/kWh  
Bậc 6: Từ 401 trở đi kWh: 3.460 đồng/kWh 
"""
ten = input("Nhập vào họ và tên chủ hộ : ")
chi_so_truoc = float(input("Nhập vào chỉ số điện kế tháng trước : "))
chi_so_sau = float(input("Nhập vào chỉ số điện kế tháng này : "))
so_kwh = chi_so_sau - chi_so_truoc
if so_kwh <=50 : # Bậc 1
    tien = so_kwh*1984
elif 50<so_kwh<=100 : # Bậc 2
    tien = 50*1984 + (so_kwh-50)*2050
elif 100<so_kwh<=200 : # Bậc 3
    tien = 50*(1984+2050) + (so_kwh-100)*2380
elif 200<so_kwh<=300 : # Bậc 4
    tien = 50*(1984+2050) + 100*2380 + (so_kwh-200)*2998
elif 300<so_kwh<=400 : # Bậc 5
    tien = 50*(1984+2050) + 100*(2380+2998) + (so_kwh-330)*3350
else : # Bậc 6 
    tien = 50*(1984+2050) + 100*(2380+2998+3350) + (so_kwh-400)*3460
print("Họ tên chủ hộ : ",ten)
thanhtoan = tien*0.08 + tien
print("Số kwh tiêu thụ trong tháng : ",so_kwh)
print("Số tiền điện tháng này : ",round(thanhtoan))
```

```python
""" 1. Kiểm tra số chẵn lẻ: Viết chương trình nhập một số nguyên, kiểm tra và in ra số đó 
là số chẵn hay số lẻ. 
"""
n = int(input("Nhập vào một số nguyên : "))
if n%2==0 :
    print(n,"là số chẵn")
else :
    print(n,"là số lẻ")
```

```python
""" 2. Kiểm tra số có tận cùng 5 không: Viết chương trình nhập một số nguyên, kiểm tra 
số đó có tận cùng 5 không, nếu đúng in ra “True”, nếu sai in ra “False”. 
"""
n = int(input("Nhập vào một số nguyên : "))
if n % 10 == 5:
    print("True")
else:
    print("False")
```

```python
""" 3. Kiểm tra số có chia hết cho 3 và 5 không: Viết chương trình nhập một số nguyên, 
kiểm tra số đó có chia hết cho 3 và 5 không, nếu đúng in ra “True”, nếu sai in ra 
“False”. 
"""
n = int(input("Nhập vào một số nguyên : "))
if n%3==0 and n%5==0 :
    print("True")
else :
    print("False")
```

```python
""" 4. Kiểm tra tuổi đủ điều kiện bầu cử: Viết chương trình nhập năm sinh của một 
người, kiểm tra xem người đó có đủ 18 tuổi để đi bầu cử hay không. 
"""
nam_sinh = int(input("Nhập vào năm sinh : "))
tuoi = 2026 - nam_sinh
if tuoi >= 18:
    print("Sinh năm",nam_sinh ,"thì người đó đủ điều kiện để đi bầu cử.")
else:
    print("Sinh năm",nam_sinh ,"thì người đó không đủ điều kiện để đi bầu cử.")
```

```python
""" 5. So sánh 2 số: Viết chương trình nhập vào hai số, in ra số lớn hơn hoặc thông báo 
nếu hai số bằng nhau. 
"""
a,b = map(float,input("Nhập vào 2 số : ").split())
if a>b or b>a :
    print("Số lớn hơn là : ",max(a,b))
else:
    print("Hai số bằng nhau.")
```

```python
""" 6. Kiểm tra ký tự nhập vào là chữ hay số: Viết chương trình nhập một ký tự, kiểm 
tra xem ký tự đó là chữ cái hay số.
"""
c = input("Nhập vào 1 ký tự : ")
if c.isalpha():
    print(c,"ký tự nhập vào là chữ cái.")
elif c.isdigit():
    print(c,"ký tự nhập vào là số.")
else:
    print(c,"ký tự nhập vào không phải là chữ cái cũng không phải là số.")
```

```python
""" 7. Bài tập 5: Tính điểm học lực 
Nhập điểm (0-10), phân loại học lực theo quy tắc: 
>=8: Giỏi; >=6.5 và <8: Khá; >=5 và <6.5: Trung bình; <5: Yếu 
"""
diem = float(input("Nhập vào điểm : "))
if diem>=8 :
    print(diem,"Học lực giỏi")
elif diem>=6.5 :
    print(diem,"Học lực khá")
elif diem>=5 :
    print(diem,"Học lực trung bình")
else:
    print(diem,"Học lực yếu")
```

```python
""" 8. Kiểm tra năm nhuận: Viết chương trình nhập năm, kiểm tra xem năm đó có phải 
năm nhuận hay không. (Năm nhuận nếu chia hết cho 400 hoặc chia hết cho 4 
nhưng không chia hết cho 100) 
"""
n = int(input("Nhập vào một năm : "))
if n%400==0 or n%4==0 and n%100!=0 :
    print(n,"là năm nhuận ")
elif n%400!=0 and n%4!=0 or n%100==0 :
    print(n,"Không phải là năm nhuận")
```

```python
""" 9. Chuyển đổi số thành chữ số: Nhập một số nguyên từ 0 đến 9, in ra tên chữ số 
tương ứng (0 -> "Không", 1 -> "Một", ...) – Xem trước match …case 
"""
n = int(input("Nhập vào một số nguyên từ 0 đến 9 : "))
chu = ["không","một","hai","ba","bốn","năm","sáu","bảy","tám","chín"]
if 0<=n<=9 :
    print("Số",n,"tương ứng với chữ số : ",chu[n])
```

```python
""" 10. Kiểm tra điểm thi xem có qua môn hay phải học lại: Nhập điểm thi, nếu điểm >= 
4 thì in "Qua môn", ngược lại in "Học lại".
"""
diem = float(input("Nhập vào điểm thi : "))
if diem>=4 :
    print(diem,"qua môn ")
else :
    print(diem,"học lại")
```
