```python
# Bài 1 : Nhập vào 2 cạnh của một hình chữ nhật. In ra màn hình diện tích và chu vi của nó. 
a,b = map(float,input("Nhập vào 2 cạnh của hình chữ nhật : ").split())
print("Hình chữ có độ dài hai cạnh a,b có diện tích là {0} và chu vi là {1}".format(a*b,2*(a+b)))
```

```python
# Bài 2: Nhập vào bán kính của hình tròn. In ra màn hình diện tích và chu vi của nó.
import math
r = float(input("Nhập vào bán kính hình tròn : "))
print("Hình tròn với bán kính có diện tích là {0} và chu vi là {1} ".format(round(math.pi*r**2, 2), round(2*math.pi*r, 2)))
```

```python
""" Bài 3 : Tính tam giác: Nhâp vào ba số a,b,c bất kì. Kiểm tra xem nó có thể là 
độ dài ba cạnh hay không, nếu  không thì in  ra màn  hình  'Khong la ba  canh 
cua mot tam  giac'. Ngược lại, kiểm tra xem nó là tam giác gì: Đều, Cân, 
Vuông, hay Tam giác thường. Tính diện tích, chu vi của tam giác và in ra màn 
hình. 
"""
a,b,c = map(float,input("Nhập vào 3 cạnh của tam giác : ").split())
if a+b>c and a+c>b and b+c>a :
    print("Có thể là 3 cạnh của tam giác ")
    p=(a+b+c)/2
    print("tam giác với 3 cạnh ",a,b,c,"có diện tích là {0} và chu vi là {1}".format(round((p*(p-a)*(p-b)*(p-c))**0.5,1),round(p*2,1)))
    if a==b==c :
        print("Đây là tam giác đều ")
    elif a==b or a==c or b==c :
        print("Đây là tam giác cân ")
    elif a**2==b**2+c**2 or b**2==a**2+c**2 or c**2==a**2+b**2 :
        print("Đây là tam giác vuông")
    else :
        print("Đây là tam giác thường")
else :
    print(a,b,c,"không là cạnh của tam giác")
```

```python
# Bài 4: Viết chương trình giải phương trình bậc 1 (ax + b = 0) 
a,b = map(float,input("Nhập vào hệ số a và b : ").split())
if a==0 and b==0 :
    print("Phương trình có vô số nghiệm ")
elif a==0 and b!=0 :
    print("Phương trình vô nghiệm ")
else :
    print("Phương trình có nghiệm duy nhất x = {0}".format(round(-b/2*a,1)))
```

```python
# Bài 5: Viết chương trình giải phương trình bậc 2 (ax2 + bx + c = 0)
import math 
a,b,c = map(float,input("Nhập hệ số cho phương trình : ").split())
if a==0 :
    if b==0 and c== 0 :
        print("Phương trình có vô số nghiệm")
    elif b==0 and c!=0 :
        print("Phương trình vô nghiệm ")
    else :
        print("Phương trình có nghiệm duy nhất x = {0}".format(round(-c/b,1)))
else :
    print(a,b,c)
    delta = b**2 - 4*a*c
    if delta < 0 :
        print("Phương trình vô nghiệm ")
    elif delta == 0 :
        print("Phương trình có nghiệm kép x = {0}".format(round(-b/(2*a),1)))
    else :
        x1 = (-b + math.sqrt(delta))/(2*a)
        x2 = (-b - math.sqrt(delta))/(2*a)
        print("Phương trình có 2 nghiệm phân biệt là x1 = {0} và x2 = {1}".format(round(x1,1),round(x2,1)))
```

```python
# Bài 6: Tìm giá trị lớn nhất trong 4 số a, b, c, d (a, b, c, được nhập từ bàn phím). 
a,b,c,d = map(float,input("Nhập vào 4 số a,b,c,d : ").split())
print("Số lớn nhất trong 4 số {0},{1},{2},{3} là {4}".format(a,b,c,d,max(a,b,c,d)))
```

```python
# Bài 7: Tìm giá trị nhỏ nhất trong 4 số a, b, c, d (a, b, c, được nhập từ bàn phím).
a,b,c,d = map(float,input("Nhập vào 4 số a,b,c,d : ").split())
print("Số nhỏ nhất trong 4 số {0},{1},{2},{3} là {4}".format(a,b,c,d,min(a,b,c,d)))
```

```python
""" Bài 8: Giải hệ phương trình tuyến tính: 
ax  by  m 
cx  dy  n
"""
a,b,m,c,d,n = map(float,input("Nhập hệ số cho hệ phương trình : ").split())
if a/c == b/d != m/n :
    print("Hệ phương trình vô nghiệm ")
elif a/c == b/d == m/n :
    print("Hệ phương trình có vô số nghiệm ")
else :
    x = (m*d - b*n)/(a*d - b*c)
    y = (a*n -m*c)/(a*d - b*c)
    print("Hệ phương trình có bộ nghiệm duy nhất là x = {0} và y = {1}".format(round(x,1),round(y,1)))
```

```python
""" Bài 9: Nhập vào thời gian 1 công việc nào đó là x giây. Hãy chuyển đổi và viết 
ra màn hình số thời gian trên dưới dạng bằng bao nhiêu giờ, bao nhiêu phút, bao 
nhiêu giây.  
"""
x = float(input("Nhập vào thời gian công việc được tính bằng giây  : "))
gio = x//3600 
phut = (x%3600)//60
giay = (x%3600)%60
print("{0}giờ {1}phút {2}giây".format(gio,phut,giay))
```

```python
""" Bài 10: Nhập vào tâm và bán kính của một đường tròn. Sau đó nhập vào một 
điểm A(x, y) bất kì và kiểm tra xem nó có thuộc đường tròn hay không? 
"""
xI,yI = map(float,input("Nhập vào tọa độ của I : ").split())
r = float(input("Nhập vào bán kính r : "))
xA,yA = map(float,input("Nhập vào tọa độ của A : ").split())
if (xA-xI)**2 + (yA-yI)**2 == r**2 :
    print("Điểm A({0},{1}) thuộc đường tròn có tâm I({2},{3}) và bán kính r = {4}".format(xA,yA,xI,yI,r))
else :
    print("Điểm A({0},{1}) không thuộc đường tròn có tâm I({2},{3}) và bán kính r = {4}".format(xA,yA,xI,yI,r))
```
