```python
""" W3A3 : Kiểm tra xem một số có phải là lũy thừa của 2 hay không 
Sử dụng các toán tử bitwise để xác định xem một số có phải là lũy thừa của 2 hay không. 
Gợi ý: Đối với lũy thừa của 2, n & (n-1) == 0
"""
n = int(input("Nhập vào 1 số n : "))
for i in range(1,n+1) :
    if n == 2**i :
        print("True")
        break
else :
    print("False")
```

```python
""" W3A11 : Nhập vào 3 số nguyên dương a,b,c là độ dài 3 cạnh của một tam giác. Kiểm tra 3 số đã nhập có tạo thành một tam giác hợp 
lệ hay không. Nếu không in ra "Không phải tam giác". Nếu có, phân loại tam giác đó là "Tam giác đều", "Tam giác cân", hay 
"Tam giác thường".
"""
a,b,c = map(float,input("Nhập vào 3 số a,b,c : ").split())
if a+b>c and a+c>b and b+c>a :
    print("Có tạo thành tam giác")
    if a==b==c :
        print("Tam giác đều")
    elif a==b or a==c or b==c :
        print("Tam giác cân")
    else :
        print("Tam giác thường")
else :
    print("Không tạo thành tam giác")
# phép thử a=3,b=4,c=5
```

```python
""" W3A12 : Nhập vào một số nguyên dương n là số năm, kiểm tra xem năm đó có phải là năm nhuận hay không. Năm nhuận là năm 
chia hết cho 4 nhưng không chia hết cho 100, hoặc chia hết cho 400. Nếu là năm nhuận in ra "Yes", nếu không in ra "No" 
"""
n = int(input("Nhập vào số năm : "))
if n%4==0 and n%100!=0 or n%400==0 :
    print("Yes")
else :
    print("No")
# Phép thử n = 2026
```

```python
""" W3A13 : [Tính tiền điện] 
Nhập vào số kWh điện tiêu thụ. Tính tiền điện theo các bậc sau: 
- 0-50 kWh: 1500đ/kWh 
- 51-100 kWh: 2000đ/kWh 
- 100 kWh: 3000đ/kWh
"""
so_kwh = float(input("Nhập vào số điện đã sử dụng : "))
if so_kwh <= 50 :
    tien = so_kwh*1500
elif 50<so_kwh<=100 :
    tien = 50*1500 + (so_kwh-50)*2000
else :
    tien = 50*(1500+2000) + (so_kwh-100)*3000
print(tien)
# phép thử so_kwh = 40
```

```python
""" W3A14 : [Giải phương trình bậc nhất] 
Nhập vào hai số thực a và b của phương trình ax + b = 0. Tìm và in ra nghiệm của phương trình, kết quả làm tròn đến số 
thập phân thứ 2. In ra "Vô nghiệm" hoặc "Vô số nghiệm" trong trường hợp tương ứng.
"""
a,b = map(float,input("Nhập vào hệ số a và b : ").split())
if a==0 and b==0 :
    print("Phương trình có vô số nghiệm ")
elif a==0 and b!=0 :
    print("Phương trình vô nghiệm ")
else :
    print("Phương trình có nghiệm duy nhất x = {0}".format(round(-b/2*a,1)))
# phép thử a=0 và b=0
```

```python
""" W3A15 : [Xếp loại học lực] 
Nhập vào điểm trung bình của một học sinh và in ra học lực của học sinh đó.Xếp loại học lực theo quy tắc: 
- >= 8.0: Giỏi 
- >= 6.5: Khá 
- >= 5.0: Trung bình 
- < 5.0: Yếu
"""
diem_tb = float(input("Nhập vào điểm trung bình : "))
if diem_tb >= 8 :
    print("giỏi")
elif 6.5 <= diem_tb < 8 :
    print("khá")
elif 5.0 <= diem_tb < 6.5 :
    print("trung bình")
else :
    print("yếu") 
# Phép thử diem_tb = 6.7
 
```

```python
""" W3A16 : Nhập vào một số thực: in ra kết quả làm tròn lên ,tròn xuống, làm tròn tới số nguyên gần nhất (không được sử dụng hàm 
sẵn có của pyton). Các kết quả cách nhau một dấu cách
"""
import math
n = float(input("Nhập số thực n : "))
print("Làm tròn lên {0},làm tròn xuống {1},lấy phần nguyên gần nhất {2}".format(math.ceil(n),math.floor(n),round(n)))
# Phép thử n = 7.4
```

```python
""" W3A17 : Cấp số nhân 
Cho 4 số a, b, c, d. Hãy kiểm tra xem 4 số này có thể theo thứ tự tạo thành 1 cấp số nhân với công bội nguyên theo đúng 
thứ tự a, b, c, d hay không?
"""
a,b,c,d = map(float,input("Nhập vào 4 số a,b,c,d : ").split())
q = b/a 
if c/b == d/c == q :
    print("Yes")
else :
    print("no")
# Phép thử a=b=c=d=92
```
