```python
""" W4A8 : Nhập số nguyên dương n. Lặp lại phép toán n = n + reverse(n) cho đến khi n là số 
đối xứng (palindrome). In ra số bước và giá trị palindrome thu được. 
"""
n = int(input("Nhập số nguyên n : "))
steps = 0
while str(n) != str(n)[::-1] :
    rev = int(str(n)[::-1])
    n = n + rev 
    steps+=1
print(steps,n)
# Phép thử n = 123
```

```python
""" W4A9 : Nhập vào số nguyên dương n. In ra tất cả các số chính phương ≤ n có các chữ số 
đôi một khác nhau (in theo thứ tự tăng dần, cùng hàng, cách nhau bởi một dấu 
cách). Nếu không có, in ra (no number). 
"""
n = int(input("Nhập vào số nguyên dương n : "))
found = False
for i in range(1,n+1) :
    if (i*i) <= n :
        print(i*i,end=" ")
        found = True
if not found :
    print("no number")
# Phép thử n = 25
```

```python
""" W4A10 : Nhập vào số nguyên dương n. Tìm số x trong [1, n] có độ dài dãy Collatz lớn nhất 
và in ra x cùng độ dài dãy tương ứng (x L). Nếu có nhiều số cùng độ dài, chọn số 
nhỏ nhất." 
"""
n = int(input("Nhập n: "))

max_len = 0
best_x = 1

for x in range(1, n+1):
    temp = x
    length = 0
    
    while temp != 1:
        if temp % 2 == 0:
            temp = temp // 2
        else:
            temp = temp * 3 + 1
        length += 1
    
    if length > max_len:
        max_len = length
        best_x = x

print(best_x, max_len)
# phép thử n=270
```

```python
# W4A11 : Viết chương trình đếm số lượng ước số chẵn của số nguyên dương n với n < 10^6.
n = int(input("Nhập số nguyên dương : "))
dem = 0
for i in range(1,n+1) :
    if n%i==0 and i%2==0 :
        dem+=1
print(dem)
        
# Phép thử n = 70
```

```python
""" W4A13 : Hai số nguyên dương tạo thành một cặp số thân thiết khi chúng tuân theo quy luật: 
Số này bằng tổng tất cả các ước của số kia (trừ chính nó) và ngược lại. 
Ví dụ 220 và 284 là một cặp số thân thiết 
220 ngoài chính nó ra, nó còn có 11 ước số khác là: 1, 2, 4, 5, 10, 11, 20, 44, 55, 
110. Tổng của 11 ước này cho chúng ta 284 
284 ngoài bản thân nó, còn có 5 ước số khác là: 1, 2, 4, 71, 142. Tổng của chúng 
vừa đúng bằng 220 
Nhập hai số từ bàn phím và kiểm tra xem chúng có phải là cặp số thân thiết không. 
Nếu có, in ra màn hình "true", ngược lại, in ra "false
"""
a,b = map(int,input("Nhập vào 2 số nguyên : ").split())
tong1 = 0
tong2 = 0
for i in range(1,a):
    if a%i==0 :
        tong1 += i
for v in range(1,b):
    if b%v==0 :
        tong2 += v
if tong1==b and tong2==a :
    print("true")
else :
    print("false")
# Phép thử giống đề bài
```

```python
# W4A14 : Cho hai số m và n là hai số nguyên dương nhập từ bàn phím. Viết chương trình 
# tìm ước chung lớn nhất của hai số m và n. 
m,n = map(int,input("Nhập vào 2 số nguyên dương : ").split())
UCLN = 0 
for i in range(1,min(m,n) +1):
    if n%i==0 and m%i==0 :
        UCLN = i
print(UCLN)
# Phép thử m,n = 2,4
```

```python
""" W4A15 : 
"Vừa gà vừa chó, 
Bó lại cho tròn, 
Ba mươi sáu con, 
Một trăm chân chẵn" 
Từ bài toán dân gian trên, mở rộng thành chương trình nhận đầu vào là tổng số 
con và tổng số chân của gà và chó. Nếu tìm được số phù hợp, in ra số lượng gà và 
số lượng chó. Ngược lại, in ra "invalid".
"""
n = int(input("Tổng số con: "))
m = int(input("Tổng số chân: "))

if (m - 2*n) % 2 != 0:
    print("invalid")
else:
    c = (m - 2*n) // 2
    g = n - c
    
    if c < 0 or g < 0:
        print("invalid")
    else:
        print(g, c)
# phép thử n,m = 10,28
```
