```python
# Bài 8: Nhập vào số nguyên dương a và b, in toàn bộ ước dương chung của a và b 
a,b = map(int,input("Nhập số nguyên dương a,b : ").split())
for i in range(1,min(a,b)+1) :
    if a%i==0 and b%i==0 :
        print(i,end=" ")
# Phép thử a=52 và b = 154
    
```

```python
""" Bài 9: Nhập vào số nguyên dương a, kiểm tra xem a có phải là số chính phương hay 
không? Số chính phương là số có căn bậc hai là một số nguyên VD: 1, 4, 9, 16, 25…. 
"""
a = int(input("Nhập số nguyên dương a : "))
for i in range(1,a+1) :
    if a==i**2 :
        print("Là số chính phương ")
        break
else :
    print("Không phải là số chính phương")
# Phép thử a=25
```

```python
""" Bài 10: Nhập vào số nguyên dương a, kiểm tra xem a có phải là số hoàn hảo hay không. 
Số hoàn hảo là số có tổng các ước (ước nhỏ hơn nó) bằng chính nó, VD: số 6, 28… 
"""
a = int(input("Nhập vào số nguyên dương a : "))
tong = 0
for i in range(1,a) :
    if a%i== 0 :
        tong += i
        if tong == a :
            print("a là số hoàn hảo ")
            break
else:
    print("a không phải là số hoàn hảo")
        
```

```python
""" Bài 11: Nhập vào số nguyên a, nếu nhập số âm thì yêu cầu nhập lại cho đến khi người 
dùng nhập đúng số dương. Nếu người dùng nhập đúng số dương thì in ra “Bạn nhập 
đúng quy tắc” và dừng chương trình. 
"""
a = int(input("Nhập vào a : "))
while a<0 :
    print("Bạn nhập sai , hãy nhập lại số dương !")
    a = int(input("Nhập lại a "))
print("Bạn nhập đúng quy tắc")
# Phép thử a=-2 và a=+2
```

```python
""" Bài 12: Nhập số nguyên dương n, Cho S(k) = 1 + 2 + 3 + … + k. Tìm k sao cho S(k) 
lớn nhất nhưng nhỏ hơn n 
"""
n = int(input("Nhập số nguyên dương n : "))
k = 0
S = 0 
while S + (k+1) < n :
    k = k + 1
    S = S + k
print("k = ",k)
print("S(k) = ",S)
# Phép thử n=6
```

```python
""" Bài 13: Nhập vào số nguyên dương A, tìm n nhỏ nhất sao cho 
1 + 1/2 + 1/3 + 1/4 + ... + 1/n > A 
"""
A = int(input("Nhập số nguyên A : "))
s = 0
n = 0
while s<=A :
    n = n+1
    s=s+1/n
print("n nhỏ nhất là : ",n) 
print("Tổng s = ",s)
```

```python
""" Bài 14: Nhập vào một dãy số nguyên, dừng nhập khi người dùng nhập -1. Sau khi nhập 
xong, in số lớn nhất, số nhỏ nhất trong những số vừa nhập. KHÔNG ĐƯỢC dùng mảng
"""
n = int(input("Nhập vào số nguyên : "))
if n == -1 :
    print("Không có dữ liệu ")
else :
    max_n = n
    min_n = n
    while n !=-1 :
        if n > max_n :
            max_n = n
        elif n < min_n :
            min_n = n
        n = int(input("Nhập số"))
print("Số lớn nhất : ",max_n)
print("Số nhỏ nhất : ",min_n)        
```

```python
""" Bài 16: Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số chẵn, bao nhiêu 
chữ số lẻ 
"""
n = int(input("Nhập vào số nguyên dương n : "))
so_chan = 0
so_le = 0
while n>0 :
    s=n%10
    if s%2==0 :
        so_chan+=1
    else :
        so_le+=1
    n=n//10
print("số chẵn trong n là : ",so_chan)
print("Số lẻ trong n là ",so_le)
# Phép thử n=1111
```

```python
# Bài 17: Nhập vào số nguyên dương n, tính tổng các chữ số của n 
n = int(input("Nhập vào số nguyên dương n : "))
tong = 0
while n>0 :
    s=n%10
    tong = tong + s
    n=n//10
print(tong)
# Phép thử n =123456
```

```python
# Bài 18: Nhập vào một số nguyên dương n, kiểm tra xem n có phải là số dạng 3^k hay không
n = int(input("Nhập vào số nguyên dương n : "))
for i in range(1,n+1) :
    if n==3**i :
        print(n,"có dạng 3 mũ k với k = {}".format(i))
        break
else :
    print(n,"không phải dạng 3 mũ k")
# Phép thử n = 3
```

```python
""" Bài 19: Dãy số fibonacci là dãy số được định nghĩa như sau: 1, 1, 2, 3, 5, 8, 13,... với số 
kế tiếp sẽ bằng tổng hai số trước đó. Nhập vào số nguyên dương A, hãy tìm số trong 
dãy số fibonacci lớn nhất nhưng không vượt quá A 
"""
A = int(input("Nhập vào số nguyên dương A"))
f1 = 1
f2 = 1
while f2<=A :
    f = f1 + f2 
    f1 = f2
    f2 = f
print("Số fibonacci lớn nhất không vượt quá A là : ",f1)
# Phép thử n = 15
```

```python
# Bài 20: Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số là số nguyên tố.  
n = int(input("Nhập vào số nguyên dương n : "))
dem = 0
while n>0 :
    s = n%10
    if s==2 or s==3 or s==5 or s==7 :
        dem+=1
    n=n//10
print("Số nguyên tố có trong n là : ",dem)
# Phép thử n = 2357
```
